```python
   # JOGO SNAKE NO PYTHON
   # UTILIZA UM ARQUIVO DE AUDIO MP3 E UM ARQUIVO DE AUDIO WAV NA PASTA DO JOGO


import pygame                             # IMPORTTA O PYGAME
from pygame.locals import *               # IMPORTA TODOS OS MODULOS DO PYGAME
from sys import exit                      # IMPORTA O METODO EXIT DA BIBLIOTECA SYS PARA SAIR DO JOGO
from random import randint                # IMPORTA O METODO RANDINT DA BIBLIOTECA RANDOM
```

```python
pygame.init()                                     # INICIALIZA O PYGAME

pygame.mixer.music.set_volume(0.25)               # CONTROLA O VOLUME DA MUSICA DO JOGO
musica_de_fundo = pygame.mixer.music.load('BoxCat Games - Inspiration.mp3') # CARREGA A MUSICA DO JOGO
pygame.mixer.music.play(-1)                       # TOCA A MUSICA DO JOGO. O (-1) FAZ TOCAR INFINITAMENTEE 
som_colisao = pygame.mixer.Sound('smw_coin.wav')  # CARREGA O SOM DA COLISAO
som_colisao.set_volume(0.5)                       # CONTROLA O VOLUME DO SOM DA COLISAO  

largura = 800                                     # DETERMINA O TAMANHO DA TELA DO JOGO
altura = 580

x_cobra = 0                                       # DETERMINA A POSICAO INICIAL DA COBRA
y_cobra = 440

velocidade = 13                                   # DETERMINA A VELOCIDADE DO JOGO
x_controle = 5                                    #  DETERMINA A VELOCIDADE INCIAL DA COBRA NOS
y_controle = 0                                      #EIXOS X E Y. NO CASO SE MOVE 5 PARA A DIREITA 

x_maca = randint(20, 780)                          # DETERMINA A POSICAO ALEATORIA DO SURGIMENTO DA                   
y_maca = randint(20, 560)                           #MACA NOS EIXOS X E Y

lista_cobra = []                                  # A LISTA DA COBRA INICIAL
comprimento_inicial = 5                           # DETERMINA O COMPRIMENTO INICIAL DA COBRA

fonte = pygame.font.SysFont('arial', 30, True, True)  # DETERMINA A FONTE DO JOGO
pontos = 0                                            # PONTUACAO COMECA COM 0
game_over = False                                     # GAME OVER COMECA COM FALSO

tela = pygame.display.set_mode((largura, altura))     # CRIA A TELA DO JOGO
relogio = pygame.time.Clock()                         # CRIA O RELOGIO DO JOGO

def aumenta_cobra(lista_cobra):                   # CRIA A FUNCAO QUE FAZ A COBRA CRESCER AO COMER A MACA
    
    for XeY in lista_cobra:                       # PARA X E Y NA LISTA COBRA
        pygame.draw.rect(tela,(0,255,0),(XeY[0],XeY[1],20,20))
        # DESENHA UM RETANGULO DO TAMANHO E POSICAO QUE A COBRA PASSOU (INDICE 0 E INDICE 1 DA LISTA COBRA)

def reiniciar_jogo():                            # CRIA A FUNCAO QWE REINICIA O JOGO
       # TORNA TODAS AS VARIAVEIS DENTRO DA FUNCAO REINICIAR JOGO GLOBAIS (AS FAZ FUNCIONAR FORA DA FUNCAO)
    global pontos, comprimento_inicial, x_cobra, y_cobra, lista_cobra, lista_cabeca, x_maca, y_maca, game_over
    
    pontos = 0
    comprimento_inicial = 5                 # RESETA TODAS AS VARIAVEIS PARA SE REINICIAR O JOGO
    x_cobra = 400
    y_cobra = 270
    lista_cobra = []
    lista_cabeca = []
    x_maca = randint(40,600)
    y_maca = randint(50,430)
    game_over = False
    
    
while True:                           # LOOP PRINCIPAL DO JOGO
    relogio.tick(20)                  # RELOGIO EM 20 FRAMES POR SEGUNDO
    tela.fill((0,0,0))                # PRENCHE A TELA DE PRETO A CADA LOOP (SEM ISSO SE VERIA O RASTRO DA COBRA)
    mensagem = f'Pontos: {pontos}'    # CRIA A MENSAGEM COM A PONTUACAO DO JOGO
    texto_formatado = fonte.render(mensagem, True, (255,255,255))
                                      # FORMATA A MENSAGEM COM  A PONTUACAO DO JOGO (O TRUE Ã‰ UM METODO
                                        #DO PYGAME QUE SUAVIZA OU NAO AS BORDAS DO TEXTO)
    for event in pygame.event.get():                    # PARA CADA EVENTO NA LISTA DE EVENTOS DO PYGAME
        
        if event.type == QUIT:                     # SE O EVENTO FOR SAIR DO JOGO 
            pygame.quit()                          # ENCERRA O PYGAME
            exit()                                 # SAI DO PROGRAMA 
            
        if event.type == KEYDOWN:                  # SE O EVENTO FOR APERTAR UMA TECLA DO TECLADO
           
            if event.key == K_a:                   # SE A TELA FOR A (PARA A ESQUERDA)
                
                if x_controle == velocidade:       # SE A COBRA ESTIVER INDO PARA A DIREITA
                    pass                           # NAO FACA NADA
                else:                              # SENAO
                    x_controle = -velocidade       # COBRA SE MOVIMENTA PARA A DIREITA (NEGATIVO NO EIXO X)
                    y_controle = 0                 # NAO SE MOVIMENTA NO EIXO Y
           
            if event.key == K_d:                   # SE A TECLA FOR D (PARA A DIREITA)
               
                if x_controle == -velocidade:      # SE A COBRA ESTIVER INDO PARA A ESQUERDA 
                    pass                           # NAO FACA NADA
                else:                              # SENAO 
                    x_controle = velocidade        # COBRA SE MOVIMENTA PARA A DIREITA (POSITIVO NO EIXO X)
                    y_controle = 0                 # NAO SE MOVIMENTA NO EIXO Y
           
            if event.key == K_w:                   # SE A TECLA FOR W (PARA CIMA)
                
                if y_controle == velocidade:       # SE A COBRA ESTIVER INDO PARA BAIXO
                    pass                           # NAO FACA NADA
                else:                              # SENAO 
                    y_controle = -velocidade       # COBRA SE MOVIEMNTA PARA CIMA (NEGATIVO NO EIXO Y)
                    x_controle = 0                 # NAO SE MOVE NO EIXO X
           
            if event.key == K_s:                   # SE A TECLA FOR S (PARA BAIXO)  
                
                if y_controle == -velocidade:      # SE A COBRA ESTIVER INDO PARA CIMA
                    pass                           # NAO FACA NADA
                else:                              # SENAO 
                    y_controle = velocidade        # COBRA SE MOVIMENTA PARA BAIXO  
                    x_controle = 0                 # NAO SE MOVE NO EIXO X
                
    x_cobra += x_controle          # COBRA SE MOVIMENTA NOS EIXOS X E Y CONFORME A TECLA APERTADA
    y_cobra += y_controle
    
       #DESENHA A COBRA E A MACA NA TELA DO JOGO ->> COR ->> CORDENADAS ->> TAMANHO
    cobra = pygame.draw.rect(tela,(0,255,0),(x_cobra,y_cobra,20,20))
    maca = pygame.draw.rect(tela,(255,0,0),(x_maca,y_maca,20,20))

    if cobra.colliderect(maca):        # SE A COBRA COLIDE A COM A MACA
        pontos += 1                    # GANHA UM PONTO
        x_maca = randint(20,780)       # GERA UMA NOVA MACA ALEATORIAMENTE 
        y_maca = randint(20,520)
        som_colisao.play()             # EXECUTA O SOM DA COLISAO
        comprimento_inicial += 1       # AUMENTA EM UMA UNIDADE O COMPRIMENTO DA COBRA
        
        if pontos >= 10 and pontos < 20:
            velocidade = 18                       # AUMENTA A VELOCIDADE DO JOGO CONFORME A PONTUACAO
        elif pontos >= 20 and pontos < 30:
            velocidade = 23
        elif pontos >= 30:
            velocidade = 28
        
    lista_cabeca = []                  # LISTA CABECA COMECA COM 0 (USADA PARA DETERMINAR SE A COBRA COLIDE COM ELA MESMA)  
    lista_cabeca.append(x_cobra)       # RECEBE A TUPLA (X, Y) DA COBRA
    lista_cabeca.append(y_cobra)
    lista_cobra.append(lista_cabeca)   # ADICIONA A LISTA CABECA NA LISTA COBRA

# SE EXITE UMA TUPLA DA LISTA COBRA IGUAL NA LISTA CABECA (SE A COBRA BATEU NELA MESMA) OU A COBRA BATEU NAS BORDAS DA TELA     
    if lista_cobra.count(lista_cabeca) > 1 or x_cobra < 0 or y_cobra < 0 or x_cobra > largura or y_cobra > altura:
       
        fonte_2 = pygame.font.SysFont('arial',20,True,True)               # FORMATA A FONTE DA REINICIALIZACAO DO JOGO
        mensagem = 'GAME OVER!!Pressione R para reiniciar ou E para sair' # CRIA A MENSAGEM DE REINICIALIZACAO DO JOGO
        texto_formatado = fonte_2.render(mensagem,True,(0,0,0))           # FORMATA A MENSAGEM DE REINICIALIZACAO DO JOGO 
        game_over = True                                                  # GAME OVER RECEBE TRUE
       
        while game_over:                                                  # ENQUANTO EM GAME OVER
            tela.fill((255,255,255))                                      # PREENCHE A TELA DE BRANCO
            fonte.render('APERTE R PARA REINICIAR',True,(0,0,0))          # RENDERIZA O TEXTO DE REINICIALIZACAO  
            
            for event in pygame.event.get():                  # PARA O EVENTO NA LISTA DE EVENTOS DO PYGAME  
               
                if event.type == KEYDOWN:                     # SE O EVENTO FOR PRESSIONAR UMA TECLA
                    
                    if event.key == K_r:                      # SE A TECLA FOR R
                        reiniciar_jogo()                      # RODA A FUNCAO DE REINICIAR JOGO
                   
                    if event.key == K_e:                      # SE A TECLA FOR E
                        pygame.quit()                         # ENCERRA O PYGAME E FECHA A JANELA
                        exit()
    
            tela.blit(texto_formatado, (180,280))             # EXIBE O TEXO DE REINICIAR O JOGO NA TELA
            pygame.display.update()                           # ATUALIZA A TELA DO JOGO

    if len(lista_cobra) > comprimento_inicial:        # SE A LISTA COBRA FOR MAIOR QUE O COMPRIMENTO INICIAL      
        del(lista_cobra[0])                           # DELETA O PRIMEIRO ITEM DA LISTA COBRA
                                                      # LEMBRANDO QUE O COPRIMENTO INICIAL AUMENTA CADA VEZ
                                                       #QUE A COBRA COME UMA MACA
    
    aumenta_cobra(lista_cobra)                   # EXECUTA A FUNCA DE AUMENTAR A COBRA NA LISTA COBRA                   
    
    tela.blit(texto_formatado,(620,30))          # EXIBE O TEXO DA PONTUACAO DO JOGO 
    pygame.display.update()                      # ATUALIZA A TELA DO JOGO
```
